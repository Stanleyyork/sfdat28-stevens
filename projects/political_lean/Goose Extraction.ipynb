```python
dict = {}
```

```python
g = Goose()
for i in pld[7500:7800].index:
    try:
        row = pld.ix[i]
        url = row.url_raw
        article = g.extract(url=url)
        dict[i] = {'url_raw': url ,'clean_text': article.cleaned_text, 'title': article.title, 'meta': article.meta_description}
        if i%100 == 0:
            print i
    except Exception as e:
      print(e)
      dict[i] = {'url_raw': url ,'clean_text': 'error', 'title': 'error', 'meta': 'error'}
      pass
```

```python
x = 0
for i in dict.keys():
    all_row = all[all.url_raw == dict[i]['url_raw']]
    all['cleaned_text'][all_row.index] = dict[i]['clean_text']
    all['title'][all_row.index] = dict[i]['title']
    all['meta_description'][all_row.index] = dict[i]['meta']
```
