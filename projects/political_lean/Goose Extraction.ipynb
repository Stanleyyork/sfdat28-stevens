```python
pld = pd.read_csv('data_11_13.csv', encoding='unicode-escape')
pld.columns = ['url_raw', 'url_clean', 'url_domain', 'ugly_text', 'issue', 'political_lean']
pld.head()
```

```python
dict = {}
```

```python
g = Goose()
for i in pld[1500:2500].index:
    try:
        row = pld.ix[i]
        url = row.url_raw
        article = g.extract(url=url)
        dict[i] = {'url_raw': url ,'clean_text': article.cleaned_text, 'title': article.title, 'meta': article.meta_description}
        if i%100 == 0:
            print i
    except Exception as e:
      print(e)
      dict[i] = {'url_raw': url ,'clean_text': 'error', 'title': 'error', 'meta': 'error'}
      pass
```

```python
import itertools
import csv

with open("output_1500_2499.csv", "wb") as f:
    w = csv.writer( f )
    columns = dict.values()[0].keys()
    w.writerow(['url_raw','clean_text', 'title', 'meta'])
    for key in dict.keys():
        text = [' '+dict[key]['clean_text']+' ']
        title = [' '+dict[key]['title']+' ']
        meta = [''+dict[key]['meta']+'']
        w.writerow([dict[key]['url_raw'], text, title, meta])
```
